```python
sal_df = pd. DataFrame({
    'Department': ['IT', 'HR' , 'IT' , 'HR', 'Finance' ],
    'Salary': [50000, 40000, 55000, 45000, 60000],
    'Bonus': [5000, 4000, 5500, 4200, 6000]
})

# Aggregate by Department
result = sal_df.groupby("Department").agg({
    "Salary": ["mean", "max"],
    "Bonus": "sum",
    "Department":"count"

})

print(result)
```

# Identify Duplicates

```python
# .duplicated() checks whether a row or value is repeated in a DataFrame.
df1 = pd.DataFrame({
    "Name": ["Alice", "Bob", "Alice", "David"],
    "City": ["Delhi", "Mumbai", "Delhi", "Chennai"]
    })
```

# pivot function

```python
df1 = pd. DataFrame({
    "Department": ["IT", "IT", "HR", "HR", "Finance"],
    "Gender": ["M", "F", "M", "F", "F"],
    "Salary": [50000, 55000, 40000, 45000, 60000]
})

df1
```

```python
pivot = pd.pivot_table(df1,values ='Salary',index = 'Department',columns = 'Gender',aggfunc='mean')
```

```python
print(pivot)
```
