#### create feature matrix (X) and output dependent coloumn vector(Y)

```python
X = df.iloc[:,2:-1]
X.head()
```

```python
Y = df.iloc[:,-1]
Y.head()
```

```python
X = df.iloc[:,2:-1].values
Y = df.iloc[:,-1].values # converts into array 
```

#### split the data

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size = .2)
```

# build models

```python
#### knn
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN = KNeighborsClassifier(n_neighbors=5)
```

```python
#train
KNN.fit(xtrain,ytrain)
#test
ypred_KNN = KNN.predict(xtest)
```

```python
#### naives bias
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
NB = GaussianNB()
```

```python
### Training te NB model
NB.fit(xtrain,ytrain)
### Testing the NB model
ypred_NB = NB.predict(xtest)
```

```python
#### DT 
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DT = DecisionTreeClassifier(criterion ="entropy",max_depth=4)
```

```python
 ### Training the DT model
DT.fit(xtrain,ytrain)
### Testing the DT model
ypred_DT=DT.predict(xtest)
```

```python
#### random forrest
```

```python
from sklearn.ensemble import RandomForestClassifier
RF=RandomForestClassifier(n_estimators=1000, criterion='entropy' ) 
### Training the model:
RF.fit(xtrain, ytrain)
### Testing the model
ypred_RF=RF.predict(xtest)
```

```python
### Use some metric to evaluate the KNN classifier 
from sklearn.metrics import confusion_matrix , accuracy_score,precision_score,recall_score,f1_score
```

```python
#### ann
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import Adam
```

```python
model=Sequential()
model.add(Dense(512, activation='relu', input_shape=(xtrain.shape[1],)))
model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense (1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer=Adam(),metrics=['accuracy'])
```

```python
ANN_clf=model.fit(xtrain, ytrain, epochs=50)
```

```python
ypred_n1 = model.predict(xtest)
```

```python
print("A] KNN classifier")
print("**************")
print("1. confusion matrix : {}" .format(confusion_matrix(ytest,ypred_KNN)))
print("2. accuracy score : {}" .format(accuracy_score(ytest,ypred_KNN)))
print("3. precesion_score : {}" .format(precision_score(ytest,ypred_KNN)))
print("4. Recall_score : {}" .format(recall_score(ytest,ypred_KNN)))
print("5.F1 score  : {}" .format(f1_score(ytest,ypred_KNN)))

print("----------------------------")
print("B] Naives Bayes Classifier")
print("**************")
print("1. confusion matrix : {}" .format(confusion_matrix(ytest,ypred_NB)))
print("2. accuracy score : {}" .format(accuracy_score(ytest,ypred_NB)))
print("3. precesion_score : {}" .format(precision_score(ytest,ypred_NB)))
print("4. Recall_score : {}" .format(recall_score(ytest,ypred_NB)))
print("5.F1 score  : {}" .format(f1_score(ytest,ypred_NB)))

print("----------------------------")
print("C] Decision Tree")
print("**************")
print("1. confusion matrix : {}" .format(confusion_matrix(ytest,ypred_DT)))
print("2. accuracy score : {}" .format(accuracy_score(ytest,ypred_DT)))
print("3. precesion_score : {}" .format(precision_score(ytest,ypred_DT)))
print("4. Recall_score : {}" .format(recall_score(ytest,ypred_DT)))
print("5.F1 score  : {}" .format(f1_score(ytest,ypred_DT)))

print("----------------------------")
print("D] Rain forrest ")
print("**************")
print("1. confusion matrix : {}" .format(confusion_matrix(ytest,ypred_RF)))
print("2. accuracy score : {}" .format(accuracy_score(ytest,ypred_RF)))
print("3. precesion_score : {}" .format(precision_score(ytest,ypred_RF)))
print("4. Recall_score : {}" .format(recall_score(ytest,ypred_RF)))
print("5.F1 score  : {}" .format(f1_score(ytest,ypred_RF)))

print("----------------------------")
print("D] Artificial Neural Network ")
print("**************")
print("1. confusion matrix : {}" .format(confusion_matrix(ytest,ypred_n1_label)))
print("2. accuracy score : {}" .format(accuracy_score(ytest,ypred_n1_label)))
print("3. precesion_score : {}" .format(precision_score(ytest,ypred_n1_label)))
print("4. Recall_score : {}" .format(recall_score(ytest,ypred_n1_label)))
print("5.F1 score  : {}" .format(f1_score(ytest,ypred_n1_label)))
```
